### Recommendation system

Recommendation systems are algorithms that provide personalized suggestions or recommendations to users based on their preferences, behaviors, or similarities with other users.

There are different types of recommendation systems: such as Collaborative Filtering, which recommends items to users based on the preferences and behaviors of similar users; or Content-based filtering which recommends items to users based on the similarity between the content or attributes of the items and the user's preferences

In this case I chose a Content-based recommendation system, which suggests to the user 10 songs similar to the song of his choice based on information about the song's sound, provided by Spotify, such as energy, acoustics and so on.

I gave the possibility through the function to choose which attributes to use for the suggestion, excluding some that do not distinguish the 'type' of song, such as popularity, key, or time_signature. 

I chose to use cosine as a measure of distance as I had no interest in the magnitude of the vectors analysed, since cosine similarity measures angular distance.

```python
import operator
from sklearn.metrics.pairwise import cosine_similarity
from sklearn import preprocessing
```

```python
def select_options():
    options = []

    # List of available options
    available_options = ['acousticness','danceability','energy','instrumentalness','liveness','loudness','speechiness','tempo','valence']

    print("Available Options:")
    for index, option in enumerate(available_options, start=1):
        print(f"{index}. {option}")

    while True:
        choice = input("Select an option (enter number, 0 to finish, 'all' for all options): ")

        # Check if the choice is 'all'
        if choice.lower() == 'all':
            options = available_options.copy()
            print("All options selected.")
            break

        # Check if the choice is a valid number
        if not choice.isdigit():
            print("Invalid input. Please enter a valid number.")
            continue

        choice = int(choice)

        # Check if the choice is within the valid range
        if choice < 0 or choice > len(available_options):
            print("Invalid option. Please enter a valid option number.")
            continue

        # If choice is 0, exit the loop
        if choice == 0:
            break

        selected_option = available_options[choice - 1]

        # Check if the option is already selected
        if selected_option in options:
            print(f"Option '{selected_option}' is already selected.")
            continue

        options.append(selected_option)
        print(f"Option '{selected_option}' selected.")

    return options
```

```python
def recommend_song_by_sound_similarity(data, song, artist):
    try:
        song_and_artist_data = data[(data['track_name']==song) & (data['artist_name']==artist)]
        similar_songs = data.copy()

        #normalize data
        sound_properties = preprocessing.normalize(similar_songs.loc[:,select_options()])

        #similarity of the song we want to all songs
        similar_songs['Similarity with song'] = cosine_similarity(sound_properties, sound_properties[song_and_artist_data.index[0],None])

        similar_songs.rename(columns= {'track_name': f'Song Similar to {song}'}, inplace =True)

        similar_songs = similar_songs.sort_values(by = 'Similarity with song', ascending=False)

        similar_songs = similar_songs[['artist_name', f'Song Similar to {song}','popularity','acousticness','danceability','instrumentalness','energy','key','liveness','loudness','speechiness','tempo','valence']]

        similar_songs.reset_index(drop=True, inplace=True)

        #return first 11 except the first, that will be the same song
        return similar_songs.iloc[1:11]

    except:
        print('This song is not included in our dataset')
        
        
```

Here we can see some example with all the variables used for the analysis:

```python
recommend_song_by_sound_similarity(data,'SICKO MODE','Travis Scott')
```

```python
recommend_song_by_sound_similarity(data,'Someone You Loved','Lewis Capaldi')
```

```python
recommend_song_by_sound_similarity(data,'Baila Baila Baila','Ozuna')
```

In this case, with the song 'Welcome to the Party' by Diplo we can see how the system picks out as the first 4 suggestions 4 different versions of the same song (remixes), this goes to show the actual similarity of the suggested songs.

```python
recommend_song_by_sound_similarity(data,'Welcome to the Party','Diplo')
```
